```python
df = pd.read_csv(r'zamowienia.csv', delimiter=';')

sprzedawcy = sorted(df["Sprzedawca"].unique())
df = df.groupby('Sprzedawca').agg({'Utarg':'sum'})

plt.pie(df['Utarg'], labels=sprzedawcy, autopct=lambda pct: "{:.1f}%".format(pct))
plt.show()
```
